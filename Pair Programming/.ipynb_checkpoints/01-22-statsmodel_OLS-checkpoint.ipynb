### Pair Problem

We'll use statsmodels to fit some (Ordinary Least Squares) linear models using random data.

First make a response y of 200 random numbers. Then generate 20 random features in an X to predict with. Fit a linear model and check the summary() output. Do you have any features with P>|t| less than 0.05? (Repeat the process until you have at least one.)

Check out the model's R-squared and Adj. R-squared. Repeat the feature generation and model fitting process with 40, 60, 80, and 100 features. What happens with R-squared and Adj. R-squared?

If you have time, automate this experimental procedure and make a plot of R-squared and Adj. R-squared against number of features for one to one hundred features.

```python
import random
import numpy as np
import statsmodels.api as sm
```

```python
random.seed(123)
```

```python
def ols(number_of_features):
    random.seed(123)
    Y = [random.random() for _ in range(200)]
    X = [[random.random() for _ in range(number_of_features)] for _ in range(200)]
    
    X = np.asarray(X)
    X = sm.add_constant(X)
    
    model = sm.OLS(Y, X)
    
    results = model.fit()
    return (results.rsquared, results.rsquared_adj)
```

```python
ols(100).summary()
```

```python
x = range(1, 101)
results = [ols(num) for num in x]

r2 = [row[0] for row in results]
r2adj = [row[1] for row in results]
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(x, r2, label = "R Squared")
plt.plot(x, r2adj, label = 'Adjusted R Squared')
plt.legend(loc='best', framealpha=0.5, prop={'size':'small'})
```

```python
%matplotlib inline
import random
import numpy as np
import statsmodels.api as sm

import matplotlib
import matplotlib.pyplot as plt

def ols(number_of_features):
    random.seed(123)
    Y = [random.random() for _ in range(200)]
    X = [[random.random() for _ in range(number_of_features)] for _ in range(200)]
    
    X = np.asarray(X)
    X = sm.add_constant(X)
    
    model = sm.OLS(Y, X)
    
    results = model.fit()
    return (results.rsquared, results.rsquared_adj)

x = range(1, 101)
results = [ols(num) for num in x]

r2 = [row[0] for row in results]
r2adj = [row[1] for row in results]

plt.plot(x, r2, label = "R Squared")
plt.plot(x, r2adj, label = 'Adjusted R Squared')
plt.legend(loc='best', framealpha=0.5, prop={'size':'small'})
```
